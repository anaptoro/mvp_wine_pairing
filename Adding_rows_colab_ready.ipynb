### Notebook to add entries to the dataset

This Colab-ready version:
- loads the CSV from the repo or directly from GitHub
- appends a new row
- saves the updated CSV in the Colab session
- lets you download the updated file


```
# Colab-friendly setup
from pathlib import Path
import pandas as pd

RAW_CSV_URL = "https://raw.githubusercontent.com/anaptoro/mvp_wine_pairing/main/data/wine_food_pairings.csv"

candidate_paths = [
    Path("data/wine_food_pairings.csv"),
    Path("/content/mvp_wine_pairing/data/wine_food_pairings.csv"),
    Path.cwd() / "data" / "wine_food_pairings.csv",
]

CSV_PATH = None
for p in candidate_paths:
    if p.exists():
        CSV_PATH = p
        break

if CSV_PATH is not None:
    df = pd.read_csv(CSV_PATH)
    print(f"Loaded local CSV: {CSV_PATH}")
else:
    df = pd.read_csv(RAW_CSV_URL)
    print("Loaded CSV from GitHub")

print(f"Rows before append: {len(df)}")
df.head()
```

```
# Add a new row
new_row = {
    "wine_type": "Sauvignon Blanc",
    "wine_category": "White",
    "food_item": "moqueca baiana",
    "food_category": "Seafood",
    "cuisine": "Brazilian",
    "quality_label": "Good",  # possible: Bad, Good, Okay
}

missing_cols = [col for col in df.columns if col not in new_row]
extra_cols = [col for col in new_row if col not in df.columns]

if extra_cols:
    raise ValueError(f"These columns are not in the CSV: {extra_cols}")

for col in missing_cols:
    new_row[col] = None

new_row_df = pd.DataFrame([new_row])[df.columns]
df = pd.concat([df, new_row_df], ignore_index=True)

print("Row added successfully.")
print(f"Rows after append: {len(df)}")
df.tail()
```

```
# Save the updated CSV inside Colab
output_dir = Path("artifacts")
output_dir.mkdir(exist_ok=True)

OUTPUT_CSV = output_dir / "wine_food_pairings_updated.csv"
df.to_csv(OUTPUT_CSV, index=False)

print(f"Saved updated CSV to: {OUTPUT_CSV.resolve()}")
```

```
# Optional: download the updated CSV
from google.colab import files
files.download(str(OUTPUT_CSV))
```
